```python
# Project:  Vending Machine

# Program execution

# Defining a function for the stocked products and regarding the purchase of items
def main():
   Stocked_Products()
   Items_Purchased()
   

# Printing opening statement
print(" \033[1;32m-----WELCOME TO 'STARS' VENDING MACHINE-----\n\t-HERE ARE STOCKED PRODUCTS:--" )

# Printing Stocked products inside the vending machines
# And including products code and prices
def Stocked_Products():
   print("\033[1;35m  Item Name:\tItem ID:\tCost: ")
   print("  Juice\t\t1\t\tAed 2.25")
   print("  Chips\t\t2\t\tAed 3.50")
   print("  Choco\t\t3\t\tAed 2.75")
   print("  Cake\t\t4\t\tAed 8.50")
   print("  Soda\t\t5\t\tAed 5.50")
   print("  Milk\t\t6\t\tAed 6.25")

# Variable for print message for input of money and the float data inserted inside a variable defined in the beginning
def Items_Purchased():
    Money = float(input("\nHow much money would you like to insert?\n-WARNING!- Only '3' Purchases at a Time: Aed "))
    if Money<2.25:
      print("--Please Start Your transaction from Beginning!!!!--\nAnd Insert more or equal to Aed 2.25!")

# Input variable for money     
    # Variable for money inserted
    Money_Inserted = Money
    
    # Variable for how many items Bought/sold
    Items_Sold = 0
    
    # List for price of items
    Items_List = [2.25, 3.50, 2.75, 8.50, 5.50, 6.25,2.50]
    
    # Variable for the money spent by the customer
    Money_Spent = 0
    
    # Variable for the total amount of money used
    Total_Amount = 0
    Item = 1


    
    # Start of While loop for the money they have input and what products they want to buy
    while Item != 0 and Money > 0 and Items_Sold < 3:
        print("You have Aed " + str(Money) + " left.")
        Item = int(input("--What item would you like to buy?--\n  ---Enter Item ID---\nAnd press '0' if you want to finish your transaction. "))
        

# Use of Conditional statements
      # If the user enter '0' than progrram will run after second 'pass' which is used as future code in this program
        if Item == 0:
            pass
   # If not selected '0' then this part will run 
      # Item ID inserted should be between 1 and 6, so writing codes for that
        elif Item < 1 or Item > 6:
            print("--Please Write a valid item ID!--.")
      # If item ID is correct than program will read the next variable regarding the ID entered 
        elif Money >= Items_List[Item-1]:
            print("Thank you for buying item "+str(Item)+".")
            Money -= Items_List[Item-1]
            Money_Spent += Items_List[Item-1]
            Total_Amount += Items_List[Item-1]
            Items_Sold += 1
        # If the product item id is 3 or less and money inserted is less than 2.25 so message will be printed
        elif Item<=3 and Money<2.25:
          print("Please insert proper amount of money!")
        ## If the product item id is more than 3 and money inserted is less than 5.50 so message will be printed
        elif Item>3 and Money>5.50:
          print("Please insert proper amount of money!")
        else:
            pass
    
# Printing Information regarding limited product as a gift
    # If user made a purchase exceeding Aed 5.50 so he will get one limited products as a gift 
    if Total_Amount>=5.5:
      print("\033[1;31m\n'For making a purchsae of more than Aed 5.50 You have won our New limited product 'Power Powder' '")
    
    # If user made a purchase exceeding Aed 8.50 so he will get two limited products as a gift 
    if Total_Amount>=8.5:
      print("'For making a purchsae of more than Aed 8.50 You have won our New limited product 'Power Snack' '")
    
    # If user made a purchase exceeding Aed 12.50 so he will get three limited products as a gift 
    if Total_Amount>=12.5:
      print("'For making a purchsae of more than Aed 12.50 You have won our New limited product 'Power Drink' '")


# Printing Information regarding the purchase  
    print("\n\033[4;37m0\033[0;43m\t--------Reciept--------")
    # For number of item sold
    print("\n\033[1;34m--  Number of items sold:           " + str(Items_Sold)+"   --")
    #For how much money the user inserted
    print("--  Amount of Money Inserted:   Aed " + str(Money_Inserted)+"   --")
    # For how much the user spent
    print("--  Amount of money spent:      Aed " + str(Total_Amount)+"   --")
    # For the remaining money
    print("--  Amount of Change:           Aed " + str(Money)+"   --")


main()
```
